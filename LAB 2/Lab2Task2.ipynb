```python
import numpy as np

def relu(x):
    return max(0, x)

inputs = [-4.0, -2.0, 0.0, 3.0, 5.0]

outputs = [relu(x) for x in inputs]

print("ReLU Outputs using Math:", outputs)
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

def relu_tensorflow(x):
    return tf.maximum(0.0, x)

inputs_tensor = tf.constant([-4.0, -2.0, 0.0, 3.0, 5.0])
outputs_tensorflow = relu_tensorflow(inputs_tensor)

print("ReLU Outputs using TensorFlow (tensor):", outputs_tensorflow.numpy())

x_values = np.linspace(-5, 5, 1000)
x_tensor = tf.constant(x_values)
y_values_tensorflow = relu_tensorflow(x_tensor).numpy()

plt.plot(x_values, y_values_tensorflow, label="ReLU (TensorFlow)", color='blue')
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.title('ReLU using TensorFlow')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()
```
