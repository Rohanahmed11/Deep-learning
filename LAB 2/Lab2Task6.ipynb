```python
import math
import numpy as np
import tensorflow as tf

def softmax_scalar(x_list):
    exp_values = [math.exp(x) for x in x_list]
    sum_exp = sum(exp_values)
    return [val / sum_exp for val in exp_values]

def softmax_tensor(x):
    return tf.nn.softmax(x)

input_values = [1.0, 2.0, 3.0]

scalar_output = softmax_scalar(input_values)

tensor_input = tf.constant(input_values, dtype=tf.float32)
tensor_output = softmax_tensor(tensor_input).numpy()

print("Softmax (Scalar):", scalar_output)
print("Softmax (TensorFlow):", tensor_output)
```

```python
import math
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

def softmax_scalar(x_list):
    exp_values = [math.exp(x) for x in x_list]
    sum_exp = sum(exp_values)
    return [val / sum_exp for val in exp_values]

def softmax_tensor(x):
    return tf.nn.softmax(x)

input_values = [1.0, 2.0, 3.0]

scalar_output = softmax_scalar(input_values)

tensor_input = tf.constant(input_values, dtype=tf.float32)
tensor_output = softmax_tensor(tensor_input).numpy()

print("Softmax (Scalar):", scalar_output)
print("Softmax (TensorFlow):", tensor_output)

plt.bar(range(len(input_values)), tensor_output, tick_label=input_values)
plt.xlabel("Input Values")
plt.ylabel("Softmax Output")
plt.title("Softmax Function Output")
plt.show()
```
