```python
import numpy as np
import math

def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))

inputs = [-1.5, 0.0, 1.0, 2.0]

outputs = [tanh(x) for x in inputs]

print("Tanh Outputs using Math:", outputs)
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

def tanh_tensorflow(x):
    return tf.tanh(x)

inputs_tensor = tf.constant([-1.5, 0.0, 1.0, 2.0])
outputs_tensorflow = tanh_tensorflow(inputs_tensor)

print("Tanh Outputs using TensorFlow (tensor):", outputs_tensorflow.numpy())

x_values = np.linspace(-5, 5, 1000)
x_tensor = tf.constant(x_values)
y_values_tensorflow = tanh_tensorflow(x_tensor).numpy()

plt.plot(x_values, y_values_tensorflow, label="Tanh (TensorFlow)", color='blue')
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.title('Tanh using TensorFlow')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()
```
