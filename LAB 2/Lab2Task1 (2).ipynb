```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

def step_function_tensorflow(x):
    return tf.where(x >= 0, tf.constant(1.0), tf.constant(0.0))

inputs_tensor = tf.constant([-3.0, -1.0, 0.0, 2.0, 5.0])
outputs_tensorflow = step_function_tensorflow(inputs_tensor)

print("Step Function Outputs using TensorFlow (tensor):", outputs_tensorflow.numpy())

x_values = np.linspace(-5, 5, 1000)
x_tensor = tf.constant(x_values)
y_values_tensorflow = step_function_tensorflow(x_tensor).numpy()

plt.plot(x_values, y_values_tensorflow, label="Step Function (TensorFlow)", color='blue')
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.title('Step Function using TensorFlow')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

def step_function(x):
    return 1 if x >= 0 else 0

inputs = [-3.0, -1.0, 0.0, 2.0, 5.0]

outputs = [step_function(x) for x in inputs]

print("Step Function Outputs:", outputs)
```
