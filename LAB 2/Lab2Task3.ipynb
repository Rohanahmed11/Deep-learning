```python
import numpy as np
import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))

inputs = [-2.0, 0.0, 1.0, 3.0]

outputs = [sigmoid(x) for x in inputs]

print("Sigmoid Outputs using Math:", outputs)
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

def sigmoid_tensorflow(x):
    return 1 / (1 + tf.exp(-x))

inputs_tensor = tf.constant([-2.0, 0.0, 1.0, 3.0])
outputs_tensorflow = sigmoid_tensorflow(inputs_tensor)

print("Sigmoid Outputs using TensorFlow (tensor):", outputs_tensorflow.numpy())

x_values = np.linspace(-5, 5, 1000)
x_tensor = tf.constant(x_values)
y_values_tensorflow = sigmoid_tensorflow(x_tensor).numpy()

plt.plot(x_values, y_values_tensorflow, label="Sigmoid (TensorFlow)", color='blue')
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.title('Sigmoid using TensorFlow')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()
```
