```python
import numpy as np

def leaky_relu(x, alpha=0.1):
    return x if x >= 0 else alpha * x

inputs = [-3.0, -1.0, 0.0, 2.0]

outputs = [leaky_relu(x) for x in inputs]

print("Leaky ReLU Outputs using Math (α=0.1):", outputs)
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

def leaky_relu_tensorflow(x, alpha=0.1):
    return tf.where(x >= 0, x, alpha * x)

inputs_tensor = tf.constant([-3.0, -1.0, 0.0, 2.0])
outputs_tensorflow = leaky_relu_tensorflow(inputs_tensor, alpha=0.1)

print("Leaky ReLU Outputs using TensorFlow (α=0.1):", outputs_tensorflow.numpy())

x_values = np.linspace(-5, 5, 1000)
x_tensor = tf.constant(x_values)
y_values_alpha_01 = leaky_relu_tensorflow(x_tensor, alpha=0.1).numpy()
y_values_alpha_03 = leaky_relu_tensorflow(x_tensor, alpha=0.3).numpy()

plt.plot(x_values, y_values_alpha_01, label="Leaky ReLU (α=0.1)", color='blue')
plt.plot(x_values, y_values_alpha_03, label="Leaky ReLU (α=0.3)", color='red', linestyle='dashed')
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.title('Leaky ReLU using TensorFlow')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.legend()
plt.show()
```
