```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread(r'C:\Users\SED\Desktop\cat1.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

brightness_value = 70  
brightened_image = cv2.convertScaleAbs(image_rgb, alpha=1, beta=brightness_value)


plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(brightened_image)
plt.title('Brightened Image')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread(r'C:\Users\SED\Desktop\cat1.jpg') 
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

gray_image = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)

sobel_x = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=17)
sobel_y = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=17)  
sobel_edges = cv2.magnitude(sobel_x, sobel_y) 

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Original Grayscale Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(sobel_edges, cmap='gray')
plt.title('Sobel Edge-detected Image')
plt.axis('off')

plt.show()
```
