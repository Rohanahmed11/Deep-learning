```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Load dataset
data = pd.read_csv('Salary_dataset.csv')

# Features and target
X = data[['YearsExperience']]
y = data['Salary']

# Train the model
model = LinearRegression()
model.fit(X, y)

# Predict using the model
y_pred = model.predict(X)

# Plotting
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', linewidth=2, label='Regression Line')
plt.title('Salary vs Experience with Outlier')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()

# Predict salary for 6 years of experience
prediction_6_years = model.predict([[6]])
print(f"Predicted salary for 6 years of experience: ${prediction_6_years[0]:.2f}")
```

```python
slope = model.coef_[0]
intercept = model.intercept_

print(f"Slope (Coefficient): {slope:.4f}")
print(f"Intercept (Bias): {intercept:.2f}")
```

```python
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
```
